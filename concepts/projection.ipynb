```python
import numpy as np 
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
#Two arrays 

a = np.array([4, 1])
b = np.array([2, 3])
```

```python
a_dot_b = np.dot(a,b)
a_dot_a = np.dot(a,a)
b_dot_a = np.dot(b,a)

print(a_dot_b, a_dot_a, b_dot_a)

#Dot product tells us how closely or strongly two vectors align, so higher the value stronger is the alignment.
# Dot product is commutative so order does not matter
```

```python
# if a.b = |a||b|cos theta ..... then a.b/|a| = |b|cos theta ---> scalar projection 
# Since you are projecting b unto a ...
# To reconstruct the vector representation and not scalar ---- [ a.b/a.a ] * a
# since we are looking for porjection of b unto a, then we need to divide by the dot product of the vector whose direction we are looking at the projection and multiply with the direction vector 'a' 

proj = (a_dot_b / a_dot_a) * a 

proj 
```

```python
plt.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy',scale=1, label='a', color="red")

plt.quiver(0,0,b[0],b[1],angles='xy', scale_units='xy',scale=1, label='b', color="green")

plt.quiver(0,0,proj[0],proj[1],angles='xy', scale_units='xy',scale=1, label='projection', color="blue")

plt.xlim(0,5)
plt.ylim(0,5)

plt.legend()
plt.show()
```
